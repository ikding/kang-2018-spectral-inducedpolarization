# Fitting Cole-Cole conductivity function with Stretched exponential

**Author**: Seogi Kang

This notebook shows how we fitted Cole-Cole conductivity function by using Streched Exponential (SE) function. 
We fit 6 Cole-Cole curves having different time constant ($\tau$) and frequency exponent ($c$) values. By illustrating that we can fit Cole-Cole curves with the SE function, we show validity of the SE function to substitute the Cole-Cole function. This notebook was used to produce Figure 2 in Kang and Oldenburg (2019). 

```python
from SimPEG import *
from SimPEG import SIP
from scipy.constants import mu_0
import numpy as np
import scipy.sparse as sp
import matplotlib.pyplot as plt
from pymatsolver import PardisoSolver
from simpegEM1D import DigFilter
from matplotlib import rcParams
%matplotlib inline
rcParams['font.size'] = 14
```

## Create Cole-Cole function in frequency-domain

```python
# Computed time channels (51 channels; 1e-4s-100s)
time = np.logspace(-4, np.log10(2.), 51)
```

```python
def ColeColeSeigel(f, sigmaInf, eta, tau, c):
    w = 2*np.pi*f
    return sigmaInf*(1 - eta/(1 + (1j*w*tau)**c))
```

## Use SimPEG's inversion module to fit Cole-Cole with SE

```python
def fit_with_se(eta0, tau0, c0):
    # frequency to time transform assuming the step-off response
    wt, tbase, omega_int = DigFilter.setFrequency(time)
    frequency = omega_int / (2*np.pi)
    siginf = 1.
    sigma = ColeColeSeigel(frequency, siginf, eta0, tau0, c0)
    sigTCole, _ = DigFilter.transFilt(sigma, wt, tbase, omega_int, time, tol=1e-12)    
    
    wires = Maps.Wires(('eta', 1), ('tau', 1), ('c', 1))
    taumap = Maps.ExpMap(nP=1)*wires.tau
    etamap = Maps.ExpMap(nP=1)*wires.eta
    cmap = Maps.ExpMap(nP=1)*wires.c
    
    survey = SIP.SEMultiSurvey(time=time, locs=np.r_[0, 0, 0], n_pulse=0)
    dtrue = -sigTCole
    survey.dobs = dtrue
    m1D = Mesh.TensorMesh([np.ones(3)])
    prob = SIP.SEMultiInvProblem(m1D, etaMap = etamap, tauMap = taumap, cMap=cmap)
    prob.pair(survey)
    m0 = np.log(np.r_[eta0, tau0, c0])
    perc = 0.02
    dmisfitpeta = DataMisfit.l2_DataMisfit(survey)
    dmisfitpeta.W = 1/(abs(survey.dobs)*perc)
    reg = Regularization.Tikhonov(m1D)

    opt = Optimization.ProjectedGNCG(maxIter = 10)
    invProb = InvProblem.BaseInvProblem(dmisfitpeta, reg, opt)
    # Create an inversion object
    target = Directives.TargetMisfit()
    betaSch = Directives.BetaSchedule(coolingFactor=1, coolingRate=1)
    opt.upper = np.log(np.r_[1., 10., 1.])
    opt.lower = np.log(np.r_[1e-5, 0.001, 1e-2])
    
    invProb.beta=0.
    inv = Inversion.BaseInversion(invProb, directiveList=[target])
    reg.mref = 0.*m0
    prob.counter = opt.counter = Utils.Counter()
    opt.LSshorten = 0.5
    opt.remember('xc')
    opt.tolX = 1e-20
    opt.tolF = 1e-20
    opt.tolG = 1e-20
    opt.eps = 1e-20
    mopt = inv.run(m0)   
    return np.exp(mopt), survey.dobs, invProb.dpred
```

## Fit six different set of Cole-Cole curves

```python
eta0, tau0, c0 = 0.1, 5., 0.3
eta1, tau1, c1 = 0.1, 5., 0.5
eta2, tau2, c2 = 0.1, 5., 0.7
eta3, tau3, c3 = 0.1, 0.5, 0.3
eta4, tau4, c4 = 0.1, 0.5, 0.5
eta5, tau5, c5 = 0.1, 0.5, 0.7

mopt0, dobs0, dpred0=fit_with_se(eta0, tau0, c0)
mopt1, dobs1, dpred1=fit_with_se(eta1, tau1, c1)
mopt2, dobs2, dpred2=fit_with_se(eta2, tau2, c2)
mopt3, dobs3, dpred3=fit_with_se(eta3, tau3, c3)
mopt4, dobs4, dpred4=fit_with_se(eta4, tau4, c4)
mopt5, dobs5, dpred5=fit_with_se(eta5, tau5, c5)
```

## What if we fix $\tau$ and just invert for $\eta$ and $c$ 
This was questioned in the revision process of the paper, and the question was 
can we use the same ($\eta$, $\tau$, $c$) values for both Cole-Cole and SE function. 
To illustrate this is not possible, we try to fit Cole-Cole curves with the fixed $\tau$ value, 
and illustrate difficulties that we faced to fit the Cole-Cole curves. 

```python
def fit_with_se(eta0, tau0, c0):
    wt, tbase, omega_int = DigFilter.setFrequency(time)
    frequency = omega_int / (2*np.pi)
    siginf = 1.
    sigma = ColeColeSeigel(frequency, siginf, eta0, tau0, c0)
    sigTCole, _ = DigFilter.transFilt(sigma, wt, tbase, omega_int, time, tol=1e-12)    
    
    wires = Maps.Wires(('eta', 1), ('c', 1))
    etamap = Maps.ExpMap(nP=1)*wires.eta
    cmap = Maps.ExpMap(nP=1)*wires.c
    
    survey = SIP.SEMultiSurvey(time=time, locs=np.r_[0, 0, 0], n_pulse=0)
    dtrue = -sigTCole
    survey.dobs = dtrue
    m1D = Mesh.TensorMesh([np.ones(2)])
    prob = SIP.SEMultiInvProblem(m1D, etaMap = etamap, cMap=cmap, tau=np.array([tau0]))
    prob.pair(survey)
    m0 = np.log(np.r_[eta0, c0])
    perc = 0.02
    dmisfitpeta = DataMisfit.l2_DataMisfit(survey)
    dmisfitpeta.W = 1/(abs(survey.dobs)*perc)
    reg = Regularization.Tikhonov(m1D)

    opt = Optimization.ProjectedGNCG(maxIter = 10)
    invProb = InvProblem.BaseInvProblem(dmisfitpeta, reg, opt)
    # Create an inversion object
    target = Directives.TargetMisfit()
    betaSch = Directives.BetaSchedule(coolingFactor=1, coolingRate=1)
    opt.upper = np.log(np.r_[1.,  1.])
    opt.lower = np.log(np.r_[1e-5, 1e-2])
    
    invProb.beta=0.
    inv = Inversion.BaseInversion(invProb, directiveList=[target])
    reg.mref = 0.*m0
    prob.counter = opt.counter = Utils.Counter()
    opt.LSshorten = 0.5
    opt.remember('xc')
    opt.tolX = 1e-20
    opt.tolF = 1e-20
    opt.tolG = 1e-20
    opt.eps = 1e-20
    # opt.lower = 0.
    mopt = inv.run(m0)   
    return np.exp(mopt), survey.dobs, invProb.dpred
```

```python
eta0, tau0, c0 = 0.1, 5., 0.3
eta1, tau1, c1 = 0.1, 5., 0.5
eta2, tau2, c2 = 0.1, 5., 0.7
eta3, tau3, c3 = 0.1, 0.5, 0.3
eta4, tau4, c4 = 0.1, 0.5, 0.5
eta5, tau5, c5 = 0.1, 0.5, 0.7

mopt0, dobs0, dpred0=fit_with_se(eta0, tau0, c0)
mopt1, dobs1, dpred1=fit_with_se(eta1, tau1, c1)
mopt2, dobs2, dpred2=fit_with_se(eta2, tau2, c2)
mopt3, dobs3, dpred3=fit_with_se(eta3, tau3, c3)
mopt4, dobs4, dpred4=fit_with_se(eta4, tau4, c4)
mopt5, dobs5, dpred5=fit_with_se(eta5, tau5, c5)
```
